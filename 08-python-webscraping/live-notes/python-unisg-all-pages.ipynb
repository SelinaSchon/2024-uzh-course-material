```python
import requests
import bs4
import pandas as pd
import time
```

```python
time.sleep(5)
```

```python
def get_page_links(url, wait_time=3):

    response = requests.get(url)
    time.sleep(wait_time)
    response.raise_for_status()

    soup = bs4.BeautifulSoup(response.text)
    pagination_div = soup.find("div", class_="g-pagination")

    links = pagination_div.find_all("a")
    link_list = ["https://www.unisg.ch" + link["href"] for link in links]
    link_list = list(set(link_list))

    return sorted(link_list)
```

```python
def extract_person_data(url, wait_time=3):

    response = requests.get(url)
    time.sleep(wait_time)
    response.raise_for_status()

    soup = bs4.BeautifulSoup(response.text)
    person_container = soup.find("div", class_="g-person-container")
    person_container_list = person_container.find_all("div", class_="e-person-content")

    data = []
    for container in person_container_list:
        name = container.find("h3").text
        website = container.find("a")["href"]
        qual = container.find("span", class_="e-person-academic").text
        data.append({
            "name": name,
            "website": website,
            "qual": qual
        })

    return pd.DataFrame(data)
```

```python
extract_person_data("https://www.unisg.ch/en/universitaet/schools/management/ueber-som/dozierende/assoziierte-professuren?current=1")
```

```python
result_list = []
page_links = get_page_links("https://www.unisg.ch/en/universitaet/schools/management/ueber-som/dozierende/assoziierte-professuren")
for link in page_links:
    result_list.append(extract_person_data(link))
```

```python
big_df = pd.concat(result_list)
big_df
```

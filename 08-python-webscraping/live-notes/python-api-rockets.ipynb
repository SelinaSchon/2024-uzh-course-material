```python
import requests

def get_previous_launches(limit=100, offset=0):

    params = {
        "format": "json",
        "limit": limit,
        "offset": offset
    }
    url = "https://ll.thespacedevs.com/2.2.0/launch/previous/"

    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    return data
```

```python
data = get_previous_launches(limit=100, offset=100)
```

```python
import pandas as pd
df = pd.DataFrame(data["results"])
df.head()
```

```python
all_results = []
for offset in range(0, 200, 100):
    data = get_previous_launches(limit=100, offset=offset)
    df = pd.DataFrame(data["results"])
    all_results.append(df)

big_df = pd.concat(all_results)
```

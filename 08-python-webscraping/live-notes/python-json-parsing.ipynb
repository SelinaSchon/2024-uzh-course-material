```python
url_2 = "https://www.covid19.admin.ch/api/data/20210908-icrm7h5q/sources/COVID19Cases_geoRegion.json"
response_2 = requests.get(url_2)
print(response_2.status_code)
response_2.raise_for_status()
data_2 = response_2.json()
```

```python
import pandas as pd
df_2 = pd.DataFrame(data_2)
df_2.head()
```

```python
df_2["geoRegion"].unique()
```

```python
df_2["date"] = pd.to_datetime(df_2["datum"])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

for kanton, sub_df in df_2.groupby("geoRegion"):
    if kanton != "CH" and kanton != "CHFL":
        ax.plot(sub_df["date"], sub_df["sumTotal"], label = kanton, alpha =0.5)

ax.legend(loc="upper left", ncol = 2)

fig.set_size_inches(10, 6)
```

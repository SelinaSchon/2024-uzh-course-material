# Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris.head()
```

```python
# Nicer scatterplot
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
# In existing axes
fig, ax = plt.subplots(1, 2)
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris,ax=ax[0])
```

```python
# Group by color
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, alpha=0.5)
```

```python
# Regression line
sns.lmplot(x='sepal_length', y='sepal_width', hue="species", data=iris)
```

```python
# Fancy plot
sns.jointplot(x='sepal_length', y='sepal_width', hue="species", data=iris)
```

```python
# Fancy plot
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind="hex")
```

```python
# Look at every continuous variable
iris_long = iris.melt(id_vars="species")
iris_long.head()
```

```python
sns.violinplot(x="variable", y="value", hue="species", data=iris_long)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
sns.violinplot(x="variable", y="value", hue="species", data=iris_long, ax=ax)
```

```python
# A quick peek at your data
sns.pairplot(iris, hue="species")
```
